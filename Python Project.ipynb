# Converted 0 and 1 values of Senior Citizen to Yes/No to make it easier to understand

```python
def convert(value):
    if value == 1:
        return 'yes'
    else:
        return 'no'

df['SeniorCitizen'] = df['SeniorCitizen'].apply(convert)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame
data = {
    'SeniorCitizen': [0, 1, 0, 1, 1, 0, 1, 0, 1, 0, 0, 1],
    'Churn': [0, 1, 0, 1, 1, 0, 1, 0, 0, 1, 0, 0]
}
df = pd.DataFrame(data)

# Calculate the counts of churned and non-churned individuals
churn_counts = df.groupby(['SeniorCitizen', 'Churn']).size().unstack(fill_value=0)

# Calculate percentage
churn_percentage = churn_counts.div(churn_counts.sum(axis=1), axis=0) * 100

# Plotting
churn_percentage.plot(kind='bar', stacked=True, figsize=(6, 4))

# Customizing the plot
plt.title('Churn Percentage by Senior Citizen Status')
plt.xlabel('Senior Citizen (No = 0, Yes = 1)')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=0)  # Rotate x-axis labels for better readability
plt.legend(title='Churn', labels=['No Churn', 'Churn'])
plt.ylim(0, 100)  # Set y-axis limits from 0 to 100

# Show the plot
plt.show()

```

# ..these are the columans and i want to create subplots from this data where each plot would be countplot



```python
columns = ['gender', 'SeniorCitizen', 'Partner', 'Dependents']
num_columns = len(columns)

fig, axes = plt.subplots(1, num_columns, figsize=(20, 5))

for ax, column in zip(axes, columns):
    sns.countplot(data=df, x=column, ax=ax, hue = 'Churn')
    ax.set_title(f'Count Plot of {column}')
    ax.set_xlabel(column)
    ax.set_ylabel('Count')

plt.tight_layout()
plt.show()
```

# the majority of customers who do not churn tend to have services like phone service, internet service (particulary DSL),
# and online Security enabled. For services like online backup, tech support and streaming tv, churn rates are noticeably higher when these servies are not used or are unavailable.
